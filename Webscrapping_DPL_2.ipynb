**Importing Beautiful Soup**

```python
from bs4 import BeautifulSoup
import requests
```

**setting up the website **

```python
url = "https://www.amazon.in/"
response = requests.get(url)
print(response)
```

```python
html = response.content
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup.title)
```

**Viewing website headers**

```python
print(soup.h1)
```

```python
print(soup.h2)
```

```python
h2tags = soup.find_all('h2')
for soups in h2tags:
  print(soups)
```

```python
for soups in h2tags:
  print(soups.string)
```
